```python
car_crashes_df = sns.load_dataset("car_crashes")
car_crashes_df.head()
x = car_crashes_df[car_crashes_df.alcohol >6]
x
alcohol_by_abbrev = x.groupby("abbrev", as_index=False).alcohol.mean()
alcohol_by_abbrev

sns.barplot(x="abbrev",y="alcohol",data=alcohol_by_abbrev)
```

# from this graph we can conclude which abbrev has most alcohol percentage

```python
sns.relplot(x='alcohol', y='total', data=car_crashes_df)
```

# we can conclude the most common alcohol percentage between 4 and 6 

# summary
